What is contained in this dataset?

```python
d.columns = ['timestamp', 'trip_id', 'station']
d.head()
d.count()
```

I added some descriptive column names, but it is certainly cryptic and not self-evident at first glance. I will refer to each row as a "report;" this is the discrete unit of information extracted from the real-time data stream.

The first column is a POSIX timestamp (which is the elapsed time in seconds from a common reference (UTC) of 00:00:00 on Thursday, January 1, 1970), which can be manipulated using the time module.

```python
import time
t_example = d.loc[0,'timestamp']
print time.ctime(t_example)
time_struct = time.localtime(t_example)
print [t for t in time_struct]
```

Next is the trip_id column, which provides a quasi-unique identifier about the specific subway trip for which data are being reported. Before I explain why this is a *quasi*-unique identifier, let's investigate its meaning.

```python
id_example = d.loc[0,'trip_id']
print id_example
origin_time = id_example.split("_")[0]
line, route = id_example.split("_")[1].split("..")
print "trip originated:",origin_time
print "subway line:", line
print "direction:",route[0]
print "subway route info:",route[1:]
```

I use python's split command to extract pieces of information from the trip_id string.

The first piece of information is the time the trip commenced, measured in seconds from midnight of that day.

The second piece of information is the subway line, which is separated by an "_" in the trip_id.

Immediately following the ".." is the direction of the trip (northbound or southbound), designated by an "N" or "S".

Finally, there is a code used internally by the MTA to designate the assigned route for the trip. I don't extract any useful information from this, but I do rely on it to assemble a unique string to identify a particular trip.

The final column contains a code for the subway stop.

Altogether then, I have a collection of time-encoded reports. Each instance collected from the MTA data stream consists of a series of reports for each train currently running. Thus, I have a record of trips currently running and the next station each train is approaching. The MTA data stream does not directly report train location, but geographic locations may be extrapolated from the data here. That is to say, train trip progress is reported in 30 second intervals, with progress denoted by the next station the train will arrive at. By monitoring this progress via the data set provided here, you can extract a good deal of information about the train trips, including train location (by matching station code with GIS data), train speed and density, and trip times.

A few notes and nuances about working with this data.

1. As I mentioned, reports consist of the next station a particular train trip is approaching at the time of the report. This is not quite the same as reporting a location at a given time. Specifically, progress along a route has to be determined retrospectively, by observing the time at which a trip's next reported stop changes. In other words, the last time (not the first time) a particular stop is reported for a particular trip should be used as the most accurate time the train was at a given location.

2. I mentioned initially that the trip_id is quasi-unique. It is unique within a given day but not across dates. That is to say, a report on June 1 may have the same trip_id as one from June 2. However, these trips can be distinguished by the timestamps. Therefore, it is posisble to assemble unique trip ids by including date information from the timestamp along with the trip_id provided in the MTA data stream.
