```python
def nice_time(t, military=True):
    """Format raw time-of-day argument in seconds"""
    hh = int(t/3600.)
    mm = int((t%3600.)/60.)
    suffix = ""
    if not military:
        suffix = "AM"
        if hh>12:
            return str(hh%12) + ":" + str(mm) + "PM"
    return str(hh) + "." + str(mm) + suffix
```

```python
# plot a day's worth of data given start time
def plotFreq(freq, startTime=None, chooseLine="4"):
    useColor = "#00CC66"
    
    if startTime:
        if startTime in freq.index:
            print "found requested startTime",startTime,"in index of size",len(freq.index)
            freq = freq.loc[startTime:,:]
            print "sliced dataframe now has index of size",len(freq.index)
        else:
            print "requested startTime",startTime,"not found in"
            print freq.index
            return None
        
    i0 = 0
    t0 = freq.index[i0]
    dt = freq.index[i0+1] - t0
    oneDay = 24. * 3600.
    nsamples = int(oneDay/dt)
    tlist = freq.index[i0:i0+nsamples]
    tstrings = [nice_time(TOD_value(t, get_TOD_reference(t))) for t in tlist]
    #tstrings = [str(TOD_value(t, get_TOD_reference(t))/3600.) for t in tlist]
    tmap = dict(zip(tstrings, np.arange(len(tstrings))))
    import time
    dd = time.ctime(t0).split()
    nice_date = dd[0] + " " + dd[1] + " " + dd[2] + " " + dd[4]
    
    #print t0,nsamples,tlist
    print len(tstrings), tstrings

    #figure(x_range=station_order, y_range=tstrings)
    #figure(x_range=station_order)
    figure()
    hold()

    ref_stops = {'4':station_codes['Utica'], '5':station_codes['Flatbush'], '6':station_codes['Brooklyn Bridge']}
    useX = []
    for sta in freq.columns:
        if not station_map.get(sta)==None:
            useX.append(station_map[sta])
    useXshift = [x+1 for x in useX]

    # slice only the index range we're working with
    freq = freq.loc[tlist,:]
    s_max = freq.max().max()
    s_amp = 150.
    print "using max size",s_max

    for t_i in range(len(tlist)):
        T = tlist[t_i]
        yvals = [tmap[tstrings[t_i]] for dummy in range(len(useX))]
        #yvals = [int(T) for dummy in range(len(useX))]
        sizes = [freq.loc[T,station_index[s]]*(s_amp/s_max) for s in useX]
        #print "sizes",sizes
        #print T,useX,yvals,sizes
        scatter(useXshift, y=yvals,
                x_range=station_order, y_range=tstrings,
             color=useColor,
             alpha=0.3,
             size=sizes)   

    from bokeh.objects import Range1d
    curplot().title = "Stops per " + str(int(dt/60.)) + " minutes by Station - " + chooseLine + " Train"
    xaxis().major_label_orientation = np.pi/4
    xaxis().axis_label = "Subway Stop"
    yaxis().axis_label = "Time of Day (" + nice_date + ")"
    #yaxis().bounds = (0,35)
    show()
    return curplot
```
